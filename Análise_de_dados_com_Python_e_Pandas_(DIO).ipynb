# Python para análise da dados (DIO)

```python
# importando a biblioteca pandas
import pandas as pd
```

```python
# ler o conjunto de dados ## o parametro informado ignora os erros no arquivo
df = pd.read_csv("/content/drive/MyDrive/dataSets/Gapminder.csv", error_bad_lines=False, sep=";")
```

```python
# visuanlizando as 5 primeiras linhas
df.head(5)
```

```python
# alterando o nome das colunas
df = df.rename(columns={
          "country": "País",
          "continent": "continente",
          "year": "Ano",
          "lifeExp": "Expectativa de vida",
          "pop": "população total",
          "gdpPercap": "PIB"
          }
          )
```

```python
# total de linhas e colunas em nosso DF
df.shape
# retornar o nome das colunas
df.columns
# retornar o tipo de dados de cada coluna
df.dtypes
# retorna as últimas linhas >> tail() --- tail de cauda
df.tail(5)
# retorna informações estatísticas
df.describe()
```

```python
# realizando filtro nos dados

# retorna valores únicos 
df["continente"].unique()
# apenas valores específicos
filtro = df["continente"] == "Oceania"
df.loc[filtro]
# agrupamento de dados
df.groupby("continente")["País"].nunique()
# qual a expectativa média de vida para cada ano
df.groupby("Ano")["Expectativa de vida"].mean()
df.groupby(["País","Ano"])["Expectativa de vida"].mean()
```
